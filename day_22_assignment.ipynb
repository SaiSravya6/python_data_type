```python
from sklearn.cluster import DBSCAN
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"Wholesale customers data.csv")
df
```

```python
from sklearn.preprocessing import StandardScaler
stscaler = StandardScaler()
X=stscaler.fit_transform(df)
X
```

```python
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize=(20,6))
dendo = sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('Dendogram')
plt.xlabel('Customer data')
plt.ylabel('Eucl Distance')
plt.show()
```

```python
len(set(dendo['color_list']))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=3)
cluster=model.fit_predict(X)
```

```python
cluster
```

```python
df
```

```python
group_num=pd.DataFrame(cluster,columns=['Group'])
group_num
```

```python
pd.concat([df,group_num],axis=1)
```

```python
cust_group_data=pd.concat([df,group_num],axis=1)
cust_group_data
```

```python
# Model evaluation
from sklearn.metrics import silhouette_score


#Check how well your model is making cluster
silhouette_score(X, cluster)
```

```python
X
```

```python
from sklearn.cluster import KMeans

wcss=[2,2,11]
for i in range (2,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(2,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
