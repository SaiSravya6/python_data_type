```python
import pandas as pd
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("pima-indians-diabetes.data.csv")
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
len(df[df.duplicated()])
```

```python
#x-y split
X=df.iloc[:,:8]
Y=df.iloc[:,8]
```

```python
seed=42
test_size=0.25
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=test_size,random_state=seed)
```

```python
#train the model
model=XGBClassifier()
model.fit(X_train,Y_train) ##75%
```

```python
#train the model
model=XGBClassifier(learning_rate=0.03)
model.fit(X_train,Y_train) ##75%
```

```python
#train the model
model=XGBClassifier(learning_rate=1)
model.fit(X_train,Y_train) ##75%
```

```python
y_pred=model.predict(X_test)
predictions=[round(value) for value in y_pred]
print(predictions)
```

```python
##accuracy
accuracy=accuracy_score(Y_test,predictions)
print("Accuracy: %.2f%%" % (accuracy*100.0))
```

```python
import pandas as pd
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("Water_Potability_Data.csv")
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
len(df[df.duplicated()])
```

```python
#x-y split
X=df.iloc[:,:9]
Y=df.iloc[:,9]
```

```python
seed=42
test_size=0.25
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=test_size,random_state=seed)
```

```python
#train the model
model=XGBClassifier()
model.fit(X_train,Y_train) ##75%
```

```python
y_pred=model.predict(X_test)
predictions=[round(value) for value in y_pred]
print(predictions)
```

```python
##accuracy
accuracy=accuracy_score(Y_test,predictions)
print("Accuracy: %.2f%%" % (accuracy*100.0))
```
